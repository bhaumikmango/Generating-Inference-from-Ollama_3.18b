```python
import pandas as pd
```

```python
df=pd.read_csv('customer_churn.csv')
df.head()
```

```python
df.describe
```

```python
df.isnull().sum()
```

```python
x=df.drop('Churn', axis='columns')
x
```

```python
y=df['Churn']
y
```

```python
x=x.drop(['Location', 'Names', 'Company', 'Onboard_date'], axis='columns')
x
```

```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
plt.scatter(x.Age, y)
plt.xlabel('Age')
plt.ylabel('Churn')
plt.show()
```

```python
from sklearn import preprocessing
scaler=preprocessing.MinMaxScaler()
x[['Total_Purchase']]=scaler.fit_transform(x[['Total_Purchase']])
plt.scatter(x['Total_Purchase'], y)
plt.xlabel('Total Purchase')
plt.ylabel('Churn')
plt.show()
```

```python
plt.scatter(x.Account_Manager, y)
plt.xlabel('Account Manager')
plt.ylabel('Churn')
plt.show()
```

```python
plt.scatter(x.Years, y)
plt.xlabel('Years')
plt.ylabel('Churn')
plt.show()
```

```python
plt.scatter(x.Num_Sites, y)
plt.xlabel('Number of sites')
plt.ylabel('Churn')
plt.show()
```

```python
dataplot = sns.heatmap(x.corr(numeric_only=True), cmap="YlGnBu", annot=True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
xt, xtt, yt, ytt = train_test_split(x, y, test_size=0.25)
```

```python
from sklearn.ensemble import RandomForestClassifier
a=[]
b=[]
for i in range(1, 25):
    clf = RandomForestClassifier(n_estimators = i)
    clf.fit(xt, yt)
    pred=clf.score(xtt, ytt)
    a.append(i)
    b.append(pred)
b=b[::-1]
```

```python
plt.plot(a, b)
plt.ylabel('Score')
plt.xlabel('No. of trees')
plt.show()
```

```python
clf = RandomForestClassifier(n_estimators = 23)
clf.fit(xt, yt)
clf.score(xtt, ytt)
```

```python
import pickle 
with open('trained_model', 'wb') as f:
    pickle.dump(clf, f)
```
